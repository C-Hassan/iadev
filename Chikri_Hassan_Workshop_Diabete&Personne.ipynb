```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
df = pd.read_csv('C:/Users/chikr/Downloads/donnees2.csv', sep=' ')
print(df)
```

```python
df['etiquette'] = df['sexe'].apply(lambda x: 2 if x == 'M' else 1 if x == 'F' else None)
```

```python
df.head()
```

```python
sexe = dict( zip (df.etiquette.unique(), df.sexe.unique()))
print( sexe )
x = df[['taille', 'poids', 'pointure']]
y = df['sexe']
```

```python
import pandas as pd
import numpy as np

# Remplacer les points "." par NaN dans les colonnes "taille" et "poids"
df['taille'] = df['taille'].replace('.', np.nan)
df['poids'] = df['poids'].replace('.', np.nan)

# Assurez-vous que les colonnes sont de type float
df['taille'] = df['taille'].astype(float)
df['poids'] = df['poids'].astype(float)

# Remplacez les NaN par des valeurs appropriées, par exemple la moyenne
mean_taille = df['taille'].mean()
mean_poids = df['poids'].mean()

df['taille'].fillna(mean_taille, inplace=True)
df['poids'].fillna(mean_poids, inplace=True)

# Maintenant, vous pouvez diviser vos données en ensembles d'entraînement et de test, puis ajuster votre modèle BernoulliNB
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

model_Bernoulli = BernoulliNB()

model_Bernoulli.fit(x_train, y_train)

prediction = model_Bernoulli.predict(x_test)
print(prediction)
```
